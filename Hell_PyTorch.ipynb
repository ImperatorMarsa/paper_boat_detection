```
import torch, time

if torch.cuda.is_available():
    device = torch.device("cuda")
    
    a=torch.rand(9022, 9033, device=device)
    b=torch.rand(9022, 9033, device=device)
    
    pom=time.time()
    d=a.mm(b.t())
    print(time.time()-pom, 'on GPU\n\n\n')
    print(d.to("cpu", torch.double))

a=torch.rand(9022, 9033)
b=torch.rand(9022, 9033)
    
pom=time.time()
d=a.mm(b.t())
print(time.time()-pom, 'on CPU\n\n\n')
print(d)
```
